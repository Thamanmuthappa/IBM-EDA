First using "countplot"                                                  
To compare Attrition to attributes with less than 10 unique values.                                          
The reason why I limited the value to 10 was as any value above the same would be very hard to analyse


```python
sns.set_style('whitegrid')
sns.countplot(x= 'Attrition', data = clean_data)
```

As we can see from the above plot, the attrition count of IBM is over 200. 

```python
sns.set_style('whitegrid')
sns.countplot(x= 'Attrition', hue= "Department",data = clean_data)
```

In the above plots I have analysed the attrition within several departments of the company.                     
As we can see,                                                                                           
The R&D department had around 1000 employees and have retained about 85% of their employees, i.e, Attrition = 15%.            
The Sales department had around 450 employees and have retained about 78% of their employees, i.e, Attrition = 22%.             
The HR department had around 50 employees and have retained about 90% of their employees, i.e, Attrition = 10%.                
                                                                                     
So, we can conclude that the department with the highest attrition was "Sales".


```python
sns.set_style('whitegrid')
sns.countplot(x= 'Attrition', hue= "Gender",data = clean_data)
```

In the above plots I have compared the attrition between the male and female employees.                     
As we can see,                                                                                           
The company had around 900 male employees and have retained about 83%, i.e, Attrition = 17%.            
The company had around 600 female employees and have retained about 75%, i.e, Attrition = 25%.             
                                                                                                                  
So, we can conclude that the more male employees were retained by the company.

However, the male to female ratio remains unchanged at 3:2 (initially 900:600, which then became 750:500)

```python
sns.set_style('whitegrid')
sns.countplot(x= 'Attrition', hue= "JobLevel",data = clean_data)
```

In the above plots I have analysed the attrition within several "Job Levels" of the company.                                  
As we can see,                                                                                           
The company had around 550 employees at level 1 and have retained about 73%, i.e, Attrition = 27%.            
The company had around 525 employees at level 2 and have retained about 90.5%, i.e, Attrition = 9.5%.

The company had around 60  employees at level 5 and have retained about 92%, i.e, Attrition = 8%.                 
                                                                                                                  
So, we can conclude that the Level with the highest attrition was "Level 1" followed by "Level 3".

```python
sns.set_style('whitegrid')
sns.countplot(x= 'Attrition', hue= "JobSatisfaction",data = clean_data)
```

In the above plots I have compared the attrition with "Job Satisfaction Levels" of various employees.                           
As we can see,                                                                                           
The company had around 280 employees at level 1 and have retained about 78.5%, i.e, Attrition = 21.5%.            
The company had around 270 employees at level 2 and have retained about 83%, i.e, Attrition = 17%.          
The company had around 420 employees at level 3 and have retained about 86%, i.e, Attrition = 14%.          
The company had around 450 employees at level 4 and have retained about 89%, i.e, Attrition = 11%.                             
                                                                                                                  
So, we can conclude that the Level with the highest attrition was "Level 1" followed by "Level 2".          
So, it is fair to say that employees with higher job satisfaction have lower tendancies to leave the job, thus reducing attrition.

NOTE:
